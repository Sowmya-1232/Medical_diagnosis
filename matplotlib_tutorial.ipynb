```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
months=['Jan','Feb','Mar','April']
sales=[250,300,400,350]
plt.bar(months, sales, color='pink')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
plt.barh(months, sales, color='orange')
plt.title('Monthly Sales (Horizontal)')
plt.xlabel('Sales')
plt.ylabel('Months')
plt.show()
```

```python
height=[150,160,175,180]
weight=[50,60,65,75]
plt.scatter(weight,height,color='green')
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [250, 300, 200, 400, 350]
plt.pie(sales, labels=months, colors=['orange']*len(months), autopct='%1.1f%%')
plt.title('Monthly Sales (Pie Chart)')
plt.show()
```
